```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [90,100,45,83]
z = [40,50,90,34]

plt.plot(x,y,label="dsa marks",color = "red", linewidth = 5, markersize = 10, marker = "*",linestyle = "--")   #'r*--' short hand for color, marker and linestyle
plt.plot(x,z,label="maths marks",color = "blue", linewidth = 5, markersize = 10, marker = "o", linestyle = "--") #'bo--' short hand for color, marker and linestyle
plt.title("Lines in the lines are linin", fontdict={"fontname":"Times New Roman", "fontsize":20},color = "green")   #first push for output
plt.xlabel("plots on x are xing", fontdict={"fontname":"Times New Roman", "fontsize":20},color = "green") #second push for output
plt.ylabel("plots on y are ying", fontdict={"fontname":"Times New Roman", "fontsize":20},color = "green")  #third push for output
plt.legend(loc = "upper right")
plt.savefig("line_plot.png")  #to save the plot in the current directory
plt.show()
```

```python
t = np.arange(0,5,0.2)
print(t)
plt.title("Multiple plots in one plot", fontdict={"fontname":"Times New Roman", "fontsize":20},color = "green")
plt.plot(t,t,'go-',t,t**2,'r--',t,t**3,'bs',t,t**4,'g^')
plt.savefig("multiple_plots.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 10, 100)
print(x)

fig, axes = plt.subplots(2, 3)

axes[0,0].plot(x, np.sin(x))
axes[0,0].set_title('Sin func')
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('y')

axes[0,1].plot(x, np.tan(x))
axes[0,1].set_title('Tan func')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y')

axes[0,2].plot(x, np.exp(x))
axes[0,2].set_title('Exp func')
axes[0,2].set_xlabel('x')
axes[0,2].set_ylabel('y')

axes[1,0].plot(x, np.sqrt(x))
axes[1,0].set_title('Sqrt func')
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('y')

axes[1,1].plot(x, np.log(x))
axes[1,1].set_title('Log func')
axes[1,1].set_xlabel('x')
axes[1,1].set_ylabel('y')

axes[1,2].plot(x, x)
axes[1,2].set_title('Linear func')
axes[1,2].set_xlabel('x')
axes[1,2].set_ylabel('y')

plt.tight_layout()
plt.show()
```
